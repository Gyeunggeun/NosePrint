```python
def create_table():
    
    import pandas as pd

    # Database credentials

    mydb = establish_connection()


    myc= mydb.cursor()         
    query_show ="show databases;"
    myc.execute(query_show) 

    my_avail_dbs=[k[0] for k in myc]

    print("You have following databases already in your server")
    print(my_avail_dbs)
    chosen_db = input("Which database do you want to choose to create a table? ")
    
    query = "use " + chosen_db
    
    myc.execute(query)
    
    new_table = input("What is the name of your new table? ")
    
    cond=True
    i=0
    res=" "
    while cond:
        
        col_name=input("Enter the " + str(i+1) + " column name: ")
        if col_name=="":
            print('You must enter the column name')
            
        col_type=input("Enter the " + str(i+1) + " column data type: ")
        
        if col_type=="":
            col_type="varchar(50)"
            
        col_cons=input("Enter the " + str(i+1) + " column constraint: ")
        if col_cons=="":
            col_cons="NULL"
        
        res=res+ col_name +" " + col_type + " " + col_cons + ","
         
        Y=input("Do you want to enter more columns? Y/N")
        if Y.lower()=="n":
            p_key= input("Enter the primary key")
            res =res + " primary key (" +  p_key + ")" 
            cond=False
            f_key=input("Enter the foreign key if any: ")
            if f_key!="":
                pkey_tab=input("Enter the table at which it is a primary key")
                pkey_col=input("Enter the column name of the reference table key")
                res=res+ ", foreign key (" + f_key +") references " +  pkey_tab + " (" + pkey_col + ")"
            
        i=i+1    
    
    print(res)
    query_table = "create table " + new_table  + " ( "  + res +   ")"
    
    print(query_table)
    
    myc.execute(query_table)
    
    print("Your table has been created")
```
