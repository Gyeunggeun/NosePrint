````python
# 라이브러리 불러오기
import os
import re
import time
import ssl
from tqdm import notebook
import requests
from urllib.request import urlretrieve
from bs4 import BeautifulSoup as bs
````

## 크롤러 구성
크롤링을 시도할 때마다 다음 4개 함수를 실행해야합니다.

````python
def get_url(start: int, end: int) -> list:
  """시작 번호부터 종료 번호까지 순회하며 강아지별 url을 생성하는 함수입니다.
  """
  urls = []
  for page in range(start, end+1):
    url = f'http://animal.gg.go.kr/adoption-animals/{page}?curPage=1&srchAnimalSe=D&srchSttus=1'
    urls.append(url)
  print(f'[info] {start}번 ~ {end}번까지 총 {len(urls)}마리의 입양정보 url을 생성했습니다.')
  return urls
````

````python
def get_img(url: str) -> dict:
  """개별 입양정보 url에서 강아지의 이름과 이미지 url을 추출하는 함수입니다.
  """
  # HTML 파싱
  response = requests.get(url)
  html = response.text
  soup = bs(html, 'html.parser')
  # 강아지 이름
  name = soup.find('span', class_='title').text
  # 강아지 이미지
  imgs = soup.find_all(alt=re.compile("사진"))

  img_dict = {'title':name, 'imgs':imgs}
  return img_dict
````

````python
def crawler(start: int, end: int, num: int, path: str, origin='https://animal.gg.go.kr'):
  begin = time.process_time()
  print("[info] 크롤링 시작")
  # -----------------------------

  # 0. HTTPS 보안 설정
  ssl._create_default_https_context = ssl._create_unverified_context

  # 1. 입양정보 url 생성
  urls = get_url(start, end)

  # 2. 입양견 이미지 url 추출
  count = num
  for url in notebook.tqdm(urls, desc='강아지 발자국 찾는 중'):
    img_dict = get_img(url)

    # 페이지 유효성 검증
    if not img_dict['title'] == '':

      # 디렉토리 생성
      count += 1
      id = f'b{str(count).zfill(3)}'               # b001
      os.makedirs(path + '/' + id, exist_ok=True)  # /content/test/b001 생성
      os.chdir(path + '/' + id + '/')              # /content/test/b001/ 내부로 이동
      save_path = os.getcwd()                      # /content/test/b001

      # 3. 입양견 이미지 저장
      # "이제 이 사진은 제껍니다"
      c = 0
      for img in img_dict['imgs']:
        src = img['src']
        img_url = origin + src

        c += 1
        urlretrieve(img_url, f'{save_path}/{id}_{c}.jpg')  # /content/test/b001/b001_1
        print(f"... {save_path}/{id}_{c}.jpg 저장 완료")

  # --------------------------------------------
  elapsed = time.process_time() - begin
  print("[info] 크롤링 종료 | 소요 시간: " + time.strftime("%H:%M:%S.{}".format(str(elapsed % 1)[2:])[:12], time.gmtime(elapsed)))
````
